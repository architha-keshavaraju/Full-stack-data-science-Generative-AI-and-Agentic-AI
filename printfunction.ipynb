# Print format method
- .format() function
- no. of variables = no. of {}

```python
n1 = 20
n2 = 30
sum = n1+n2
print('The addition of {} and {} is {}'.format(n1,n2,sum))
```

```python
n1 = 20
n2 = 30
n3 = 40
sum = n1+n2+n3
print('The addition of {},{} and {} is {}'.format(n1,n2,n3,sum))
```

# f string method
- variable should ne in curly braces
- write everything inside quotes ''
- at starting simply add f

```python
n1 = 2
n2 = 3
sum = n1+n2
print(f'The addition of {n1} and {n2} is {sum}')
```
